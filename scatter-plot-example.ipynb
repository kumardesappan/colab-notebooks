```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


x      = [120, 431,  658,  469,  154,  372,  383, 383, 500]
y      = [60, 69.9, 71.7, 71.5, 75.9, 75.3, 74.9,74.9, 80]
colors = [0, 0,    10,   20,   30,   40,   50,  60, 0]
sizes =  [1,1958, 936, 354, 1581, 745, 369, 950, 1]
names =  ["", "resNet18", "mobileNetV1", "mobileNetV2", "resNet50V1p5", "regNetX-800mf", "efficientNet-lite0", "mobileNet-edgeTPU", ""]



plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')

for i, txt in enumerate(names):
    plt.annotate(txt, (x[i], y[i]), color='red')

```
